```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import MaxAbsScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,log_loss,cohen_kappa_score,RocCurveDisplay
warnings.filterwarnings('ignore')
```

```python
# DATA PREPROCESSING
data=pd.read_csv("parkinsons.csv")
data=data.drop(['name'],axis=1)
x=data[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']]
y=data['status']

param_grid={
'loss':['log_loss'],
'learning_rate':[0.1,0.2,0.5,0.7,1.0],
'max_depth':[None,1,2,3,4,5,8],
'scoring':['loss'],
'tol':[1e-01,1e-03,1e-05,1e-07]
}

dict1={
    "k_value":[],
    "best_estimator":[],
    "best_score":[],
    "Accuracy_train":[],
    "Accuracy_test":[]
}
#X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.10,random_state=1)
scaler=MaxAbsScaler()
model=HistGradientBoostingClassifier()
gscv=GridSearchCV(model,param_grid=param_grid,cv=5,n_jobs=-1)
```

```python
for i in range(1,23,1):
    fvalue_Best = SelectKBest(f_classif,k=i)
    X_kbest = fvalue_Best.fit_transform(x,y)
    X_train,X_test,Y_train,Y_test=train_test_split(X_kbest,y,test_size=0.30,random_state=0)
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    gscv.fit(X_train,Y_train)
    dict1["k_value"].append(i)
    dict1["best_estimator"].append(gscv.best_estimator_)
    dict1["best_score"].append(gscv.best_score_)
    model=gscv.best_estimator_
    model.fit(X_train,Y_train)
    dict1["Accuracy_train"].append(model.score(X_train,Y_train)),
    dict1["Accuracy_test"].append(model.score(X_test,Y_test))
    print("k=",i)
```

```python
df1=pd.DataFrame(dict1)
```

```python
#BEST MODEL FROM SUBSET OF FEATURES AND HYPER-PARAMATER TUNNING
acc_train=df1[df1["Accuracy_train"]==1.00].index
acc_test=df1[df1["Accuracy_test"]>=0.96].index
best = [value for value in acc_train if value in acc_test]
dict2={
    "no_of_features":[],
    "best_parameters":[]
}

for i in range(0,len(best),1):
    dict2["no_of_features"].append(best[i]+1)
    dict2["best_parameters"].append(dict1["best_estimator"][best[i]])

df2=pd.DataFrame(dict2)
df2
```
